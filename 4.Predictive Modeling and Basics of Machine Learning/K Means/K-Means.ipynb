```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
%matplotlib inline 
from sklearn.cluster import KMeans
```

```python
data=pd.read_csv("student_evaluation.csv")
```

```python
data.shape
```

```python
data.head()
```

```python
pd.isnull(data).sum()
```

```python
data.describe()
```

```python
kmeans = KMeans(n_clusters=2)
```

```python
kmeans.fit(data)
```

```python
pred=kmeans.predict(data)
```

```python
pred
```

```python
pd.Series(pred).value_counts()
```

```python
# inertia_ is the distance between points to the centroid
# lower the inertia_ value higher the quality
kmeans.inertia_
```

```python
kmeans.score(data)
```

```python
SSE = [] # to store inertia values
```

```python
for cluster in range(1,20):
    kmeans = KMeans(n_jobs = -1, n_clusters = cluster)
    kmeans.fit(data)
    SSE.append(kmeans.inertia_)
```

```python
len(SSE)
```

```python
frame = pd.DataFrame({'Cluster':range(1,20), 'SSE':SSE})
```

```python
# no. of clusters is optimised by elbow curve (care has to be taken that it may also depend use case and purpose)
# elbow method to indentify no. of clusters
plt.figure(figsize=(12,6))
plt.plot(frame['Cluster'], frame['SSE'], marker='o')
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

data_scaled = scaler.fit_transform(data)

# used to scale data such that mean is 0 and standard deviation is 1.
```

```python
SSE_scaled = []
```

```python
len(SSE_scaled)
```

```python
frame_scaled = pd.DataFrame({'Cluster':range(1,20), 'SSE':SSE_scaled})
plt.plot(frame_scaled['Cluster'], frame_scaled['SSE'], marker='o')
plt.xlabel("Clusters")
plt.ylabel("SSE")
```
