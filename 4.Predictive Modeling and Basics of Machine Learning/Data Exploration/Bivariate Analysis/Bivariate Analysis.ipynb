```python
#importing pandas

import pandas as pd
import matplotlib.pyplot as plt
```

```python
#reading the data file 

file=pd.read_csv("titanic.csv")
```

```python
#first few rows of titanic dataset

file.head()
```

```python
file.dtypes
```

# Continuous - Continuous Bivariate Analysis

```python
file.plot.scatter('Age','Fare')
```

```python
plt.scatter('Age','Fare',data = file)
```

```python
plt.scatter(file['Age'],file['Fare'])
```

```python
file.corr()
```

```python
file['Age'].corr(file['Fare'])
```

# Categorical - Continuous Bivariate Analysis

```python
file.groupby('Sex').groups
```

```python
file.groupby('Sex').plot.bar()
```

```python
file.groupby("Sex").get_group('male')
```

```python
file.groupby("Sex").get_group('female')
```

```python
filt = file["Sex"] == 'male'
file[filt]
```

```python
file.groupby('Sex')['Age'].groups
```

```python
file.groupby('Sex')['Age'].mean()
```

```python
file.groupby('Sex')['Age'].mean().plot.bar()
```

```python
#importin the scipy library for ttest

from scipy.stats import ttest_ind
```

```python
males=file[file['Sex']=='male']
```

```python
female=file[file['Sex']=='female']
```

```python
ttest_ind(males['Age'],female['Age'],nan_policy='omit')
```

# Categorical - Categorical Bivariate Analysis

```python
pd.crosstab(file['Sex'],file['Survived'])
```

```python
from scipy.stats import chi2_contingency
```

```python
chi2_contingency(pd.crosstab(file['Sex'],file['Survived']))
```
